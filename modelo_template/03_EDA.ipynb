```python
size_counts = df_pizzas['size'].value_counts()
size_counts
```

```python
plt.figure(figsize=(10, 10))
plt.gcf().set_facecolor('white')
plt.pie(size_counts, labels=size_counts.index, autopct='%1.1f%%', startangle=120)
plt.title('DistribuciÃ³n de tamaÃ±os de pizzas')
plt.show()
```

```python
category_counts = df_pizzas['category'].value_counts()
category_counts
```

```python
plt.figure(figsize=(10, 10))
plt.gcf().set_facecolor('white')
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('DistribuciÃ³n de categorÃ­as de pizzas')
plt.show()
```
