```python
#Importaci√≥n de las librerias necesarias a utilizar
import pandas as pd
```

```python
#Cargar los archivos CSV
order_details = pd.read_csv("C:/Users/Facu/OneDrive/Escritorio/DataScience/Proyecto Final II/Dataset Pizza Planeta/order_details.csv")
orders = pd.read_csv("C:/Users/Facu/OneDrive/Escritorio/DataScience/Proyecto Final II/Dataset Pizza Planeta/orders.csv")
pizza_types = pd.read_csv("C:/Users/Facu/OneDrive/Escritorio/DataScience/Proyecto Final II/Dataset Pizza Planeta/pizza_types.csv")
pizzas = pd.read_csv("C:/Users/Facu/OneDrive/Escritorio/DataScience/Proyecto Final II/Dataset Pizza Planeta/pizzas.csv")
```

```python
# Unimos las tablas para mayor comodidad en el manejo de los datos
pizzas_df = pd.merge(order_details, orders, on='order_id', how='left')
pizzas_df = pd.merge(pizzas_df, pizzas, on='pizza_id', how='left')
pizzas_df = pd.merge(pizzas_df, pizza_types, on='pizza_type_id', how='left')
```

```python
# Agrego para mayor analisis la columna de dia de la semana

#Convertir la columna 'date' a un formato de fecha si a√∫n no lo est√°
pizzas_df['date'] = pd.to_datetime(pizzas_df['date'], format='%Y-%m-%d')

# Agregar la columna 'day_of_week' con el nombre del d√≠a de la semana
pizzas_df['day_of_week'] = pizzas_df['date'].dt.day_name()

# Verificar los primeros registros para ver la nueva columna
print(pizzas_df[['date', 'day_of_week']].head())

```

```python
pizzas_df['day_of_week'].value_counts()	
```

```python
# Agrego para mayor analisis la de franja horaria de la venta
# Tomando como parametro Ma√±ana(06:00 - 12:00), Tarde (12:00 - 18:00), Noche(18:00 - 00:00) y Madrugada ( 00:00 - 06:00)

# Aseguro de que la columna 'time' est√© en formato de hora
pizzas_df['time'] = pd.to_datetime(pizzas_df['time'], format='%H:%M:%S').dt.time

# Funci√≥n para asignar franjas horarias
def categorize_time(hour):
    if 6 <= hour < 12:
        return 'Ma√±ana'
    elif 12 <= hour < 18:
        return 'Tarde'
    elif 18 <= hour < 24:
        return 'Noche'
    else:
        return 'Madrugada'

# Agregar la columna 'time' con solo la hora
pizzas_df['hour'] = pd.to_datetime(pizzas_df['time'], format='%H:%M:%S').dt.hour

# Aplicar la funci√≥n para crear la columna 'franja_horaria'
pizzas_df['franja_horaria'] = pizzas_df['hour'].apply(categorize_time)

# Verificar los primeros registros para ver la nueva columna
print(pizzas_df[['time', 'franja_horaria']].head())
```
