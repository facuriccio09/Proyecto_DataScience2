## Pizzeria Pizza Planeta

### **INTRODUCCION**
En un mundo donde las decisiones se basan cada vez más en datos, entender los patrones de consumo se vuelve esencial para cualquier negocio. 
En este análisis, nos sumergimos en un dataset que captura las ventas de una pizzería llamada "Pizza Planeta", desglosando cada rebanada de información para descubrir qué impulsa las decisiones de los clientes.
Este proyecto de análisis exploratorio de datos (EDA) tiene como objetivo analizar los patrones de venta de una pizzería, revelando insights clave sobre las preferencias del consumidor, las horas de mayor demanda y los productos más exitosos. A través de las herramientas de Data Science, este análisis ofrece soluciones prácticas para mejorar la eficiencia y aumentar los ingresos en un entorno competitivo.

### **CONTEXTO EMPRESARIAL**
El mercado de la comida rápida ha evolucionado significativamente, impulsado por cambios en el comportamiento del consumidor, la tecnología y la globalización. En este contexto, las pizzerías deben enfrentar varios desafíos, tales como:

1) Competencia intensa: Con muchas opciones disponibles, los consumidores son cada vez más selectivos, lo que obliga a los negocios a diferenciarse.
2) Variabilidad en la demanda: Las ventas pueden fluctuar considerablemente según la hora del día, el día de la semana y factores estacionales, lo que hace que la gestión de inventario y personal sea crucial.
3) Preferencias cambiantes: Los consumidores buscan cada vez más opciones innovadoras y saludables, lo que requiere que las pizzerías adapten sus menús a las tendencias del mercado.


### **PROBLEMA COMERCIAL**
En un contexto empresarial, la pizzería enfrenta varias preguntas importantes:

1) ¿Qué tipos de pizza son los preferidos por los clientes?
2) ¿Qué tamaños de pizza se venden mejor en diferentes franjas horarias?
3) ¿Cómo se distribuyen las ventas a lo largo de la semana y en diferentes horas del día?

Responder a estas preguntas es esencial para optimizar las decisiones de stock, marketing y personal. La falta de comprensión de estos patrones puede llevar a ineficiencias como desperdicio de alimentos, tiempos de espera prolongados o una experiencia subóptima para el cliente. Este análisis busca proporcionar una solución basada en datos para abordar estos problemas comerciales, mejorando tanto la operación como la satisfacción del cliente.

### **OBJETIVOS DEL PROYECTO**
Como objetivos principales están determinados:
1) **Identificar patrones de ventas:** Analizar como los dias de la semana y las franjas horarias afectan a las ventas de las pizzas.
2) **Explorar las preferencias de los productos:** Determinar que tipos de pizzas son las mas populares y como su tamaño influyen en las decisiones de quienes las compran
3) **Visualizar datos significativos:** Utilizar visualizaciones para contar la historia detras de los numeros y facilitar la toma de decisiones
4) **Optimizar la operación:** Utilizar los datos de ventas para mejorar la eficiencia de la cadena de suministro, reduciendo el desperdicio y maximizando las ganancias.

```python
size_counts = df_pizzas['size'].value_counts()
size_counts
```

```python
plt.figure(figsize=(10, 10))
plt.gcf().set_facecolor('white')
plt.pie(size_counts, labels=size_counts.index, autopct='%1.1f%%', startangle=120)
plt.title('Distribución de tamaños de pizzas')
plt.show()
```

Al notar que tanto el tamaño XXL y XL son minimos a comparación de S, M y L, procedemos a eliminarlos 

```python
# Identifico las filas donde el tamaño es XL o XXL
filas_con_XL_XXL = df_pizzas[df_pizzas['size'].isin(['XL', 'XXL'])].index

# Elimino las filas donde el tamaño es XL o XXL
df_pizzas_filtrado = df_pizzas.drop(filas_con_XL_XXL)

# Verifico que se hayan eliminado las filas
df_pizzas_filtrado['size'].value_counts()
```

Nuevamente realizo el grafico de la distribución segun le tamaño de pizzas

```python
size_counts = df_pizzas_filtrado['size'].value_counts()
size_counts
```

```python
plt.figure(figsize=(10, 10))
plt.gcf().set_facecolor('white')
plt.pie(size_counts, labels=size_counts.index, autopct='%1.1f%%', startangle=120)
plt.title('Distribución de tamaños de pizzas')
plt.show()
```

```python
category_counts = df_pizzas_filtrado['category'].value_counts()
category_counts
```

```python
plt.figure(figsize=(10, 10))
plt.gcf().set_facecolor('white')
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribución de categorías de pizzas')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.countplot(x='franja_horaria', data=df_pizzas_filtrado, order=['Madrugada', 'Mañana', 'Tarde', 'Noche'])
plt.title('Distribución de Ventas por Franja Horaria')
plt.xlabel('Franja Horaria')
plt.ylabel('Cantidad de Órdenes')
plt.show()
```

Al notar que no se realizan ventas en madrugada, procedo a eliminar todas aquellas que fueron vendidas en ese momento (asumiendo que es una pizzeria la cual no trabaja 24 hs)

```python
# Espacio para eliminar las de madrugada
df_pizzas_filtrado_v2 = df_pizzas_filtrado[df_pizzas_filtrado['franja_horaria'] != 'Madrugada']

#verifico que se hayan eliminado las filas
df_pizzas_filtrado_v2['franja_horaria'].value_counts()
```

```python
plt.figure(figsize=(10, 10))
sns.countplot(x='franja_horaria', data=df_pizzas_filtrado_v2, order=['Mañana', 'Tarde', 'Noche'])
plt.title('Distribución de Ventas por Franja Horaria')
plt.xlabel('Franja Horaria')
plt.ylabel('Cantidad de Órdenes')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.countplot(x='franja_horaria', data=df_pizzas_filtrado_v2, order=['Mañana', 'Tarde', 'Noche'], hue='category')
plt.title('Distribución de Ventas por Franja Horaria definida por Categoría')
plt.xlabel('Franja Horaria')
plt.ylabel('Cantidad de Órdenes')
plt.show()
```

## **CONCLUSIÓN**
El análisis exploratorio de datos (EDA) realizado sobre el conjunto de ventas de la pizzería ha revelado patrones significativos que pueden guiar la estrategia comercial de manera efectiva. 
A través de la visualización de datos y el análisis estadístico, se identificó que las ventas son notablemente más altas durante las horas de la noche y la tarde, mientras que las ventas en la mañana son escasas. Esta información sugiere la necesidad de ajustar la programación del personal y las ofertas promocionales para capitalizar las horas pico de venta.

Además, la exploración de las preferencias de los clientes mostró que los tipos de pizza que combinan sabores clásicos con opciones innovadoras son los más solicitados, lo que indica que los consumidores están dispuestos a experimentar, siempre que se mantenga un equilibrio con los sabores familiares, aunque siendo una diferencia (poco notoria) predominando el tipo Clasico (classic).
Hablando de los tipos tambien es de vital importancia hablar de los tamaños ya que el tamaño L (mediano) es el más solicitado, mostrando que por más que existan las opciones no convencionales, la clientela siempre se inclina por la media o lo standard.

En conclusión, el EDA no solo ha proporcionado una comprensión más clara del rendimiento de la pizzería, sino que también ha puesto de manifiesto áreas de oportunidad para mejorar las operaciones y la experiencia del cliente. Estos insights servirán como guía para futuras acciones y estrategias, reforzando la importancia de la ciencia de los datos en la toma de decisiones informadas en el ámbito empresarial.
